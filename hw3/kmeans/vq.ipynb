Run your k-means algorithm on `kmeans_data.mat`. Note that the lines indicate the iterations of centroids.

```python
from kmeans import kmeans
from kmeans_plot import kmeans_plot
x = sio.loadmat('kmeans_data.mat')['X']
k = 2  # modify if you want
idx1, ctrs1, iter_ctrs1 = kmeans(x, k)
idx2 = idx1
ctrs2 = ctrs1
iter_ctrs2 = iter_ctrs1
sd1 = 0
for i in range(k):
    sd1 += np.sum(np.linalg.norm(x[idx1 == i] - ctrs1[i]))
sd2 = sd1

sd = 0
for i in range(999):
    idx, ctrs, iter_ctrs = kmeans(x, k)
    for j in range(k):
        sd += np.sum(np.linalg.norm(x[idx == j] - ctrs[j]))
    if sd > sd1:
        sd1 = sd
        idx1 = idx
        ctrs1 = ctrs
        iter_ctrs1 = iter_ctrs
    if sd < sd2:
        sd2 = sd
        idx2 = idx
        ctrs2 = ctrs
        iter_ctrs2 = iter_ctrs

kmeans_plot(x, idx1, ctrs1, iter_ctrs1)
kmeans_plot(x, idx2, ctrs2, iter_ctrs2)
```

# Vector quantization
Compress images with K set to 8, 16, 32 and 64.

```python
from kmeans import kmeans
from matplotlib.pyplot import imread
img = imread('sample0.jpg')
fea = img.reshape((img.shape[0]*img.shape[1], 3)).astype(np.float64)
# YOUR (SEVERAL LINES) CODE HERE
# begin answer

k = 8
idx, ctrs, _ = kmeans(fea, k)
for i in range(k):
    fea[idx == i] = ctrs[i]

# end answer

plt.imshow(fea.reshape(img.shape).astype(np.uint8))
```
